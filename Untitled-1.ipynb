```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jessicali9530/celeba-dataset")

print("Path to dataset files:", path)
```

```python
import os
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from PIL import Image

class CelebADataset(Dataset):
    def __init__(self, root_dir, image_size=64, transform=None):
        self.root_dir = root_dir
        self.image_files = sorted(os.listdir(root_dir))  # 이미지 파일 목록 가져오기
        self.transform = transform if transform else transforms.Compose([
            transforms.Resize((image_size, image_size)),  # 이미지 크기 조정
            transforms.ToTensor(),  # (H, W, C) → (C, H, W) 변환
            transforms.Normalize(mean=[0.5], std=[0.5])  # [-1, 1] 정규화
        ])

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.root_dir, self.image_files[idx])
        image = Image.open(img_path).convert("RGB")  # RGB로 변환
        image = self.transform(image)  # 변환 적용
        return image  # 라벨 없이 이미지만 반환



# KaggleHub에서 받은 CelebA 데이터셋 경로 설정
celeba_root = os.path.join(path, "img_align_celeba/img_align_celeba")  # 이미지 폴더 경로

# PyTorch Dataset 생성
dataset = CelebADataset(celeba_root, image_size=64)

# DataLoader 생성
dataloader = DataLoader(dataset, batch_size=16, shuffle=True, num_workers=2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

im = Image.open(celeba_root + "/000001.jpg")
width, height = im.size
print(width, height)

img_np = np.array(im) ## 행렬로 변환된 이미지
plt.imshow(img_np) ## 행렬 이미지를 다시 이미지로 변경해 디스플레이
plt.show() ## 이미지 인터프린터에 출력

```

```python
# 데이터 확인
for batch in dataloader:
    print("Batch Shape:", batch.shape)  # (batch_size, C, H, W)
    break
```

```python
import torch
from denoising_diffusion_pytorch import Unet, GaussianDiffusion, Trainer

model = Unet(
    dim = 64,
    dim_mults = (1, 2, 4, 8),
    flash_attn = True
)

diffusion = GaussianDiffusion(
    model,
    image_size = 64,
    timesteps = 1000    # number of steps
)


trainer = Trainer(
    diffusion,
    os.path.join(path, "img_align_celeba/img_align_celeba"),
    train_batch_size = 64,
    train_lr = 8e-5,
    train_num_steps = 700000,         # total training steps
    gradient_accumulate_every = 2,    # gradient accumulation steps
    ema_decay = 0.995,                # exponential moving average decay
    amp = True,                       # turn on mixed precision
    calculate_fid = True              # whether to calculate fid during training
)

trainer.train()

```
